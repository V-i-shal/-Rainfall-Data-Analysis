```python
#RAINFALL PREDICTION 
import pandas as pd 
df=pd.read_csv("C:/Users/visha/OneDrive/Desktop/rainfall.csv")
print(df.shape)

df = df[['YEAR', 'ANNUAL']].dropna()
df['YEAR'] = pd.to_numeric(df['YEAR'],errors='coerce')
df['ANNUAL'] = pd.to_numeric(df['ANNUAL'],errors='coerce')
df.dropna()

import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
correlation = df['YEAR'].corr(df['ANNUAL'])
print(correlation)

x=df['YEAR'].values.reshape(-1, 1)
y=df['ANNUAL']

model=LinearRegression()
model.fit(x,y)
predictions=model.predict(x)
plt.plot(x,predictions)
plt.scatter(x,y)

#FUTURE PREDICTION 
model.predict([[2026]])
```
